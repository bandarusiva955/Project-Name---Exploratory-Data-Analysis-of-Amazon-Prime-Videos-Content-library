# Data wrangling

```python
# import libraries 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import warnings
warnings.filterwarnings('ignore')
```

```python
# fetch dataset 1
data1= pd.read_csv('titles.csv.zip')
```

```python
# fetch dataset 2
data2=pd.read_csv('credits.csv (1).zip')
```

```python
data1.columns 
```

```python
data2.columns
```

```python
# merging 2 datasets on common column id 
data=pd.merge(data1,data2,on='id')
```

```python
#  print the merged data which we use for data data analysis
data
```

```python
# dataset rows and columns count
data.shape
```

```python
# fetching all the columns of dataset
data.columns
```

```python
# dataset information 
data.info()
```

# dataset description

# missing values/null values

```python
# check for nulll values 
data.isnull().sum()
```

```python
# drop null values in description column 
data.dropna(subset = ['description'],inplace= True )
```

```python
# mode of age_certifications
data['age_certification'].mode()[0]
```

```python
#replace all null values of age_certification with mode
data['age_certification'].fillna(data['age_certification'].mode()[0],inplace= True)
```

```python
data.isnull().sum()
```

```python
# replacing null values in seasoms column with 0 
data['seasons'].fillna(0,inplace= True)
```

```python
data.isnull().sum()
```

```python
data.dropna(subset= ['imdb_id'],inplace=True)
```

```python
data.isnull().sum()
```

```python
round(data['imdb_score'].mean(),1)
```

```python
#replace all null values in imdb_score with mean of imdb_score
data['imdb_score'].fillna(round(data['imdb_score'].mean(),1),inplace=True)
```

```python
data.shape
```

```python
data['imdb_votes'].fillna(0,inplace=True)
```

```python
data.isnull().sum()
```

```python
# mean of tmdb_popularity
round(data['tmdb_popularity'].mean(),2)
```

```python
# replace null values in ' tmdb_popularity' column with meann
data['tmdb_popularity'].fillna(round(data['tmdb_popularity'].mean(),2),inplace=True)
```

```python
# mean of tmdb_score
round(data['tmdb_score'].mean(),1)
```

```python
#replace null valuess in"tmdb_score" column with mean 
data['tmdb_score'].fillna(round(data['tmdb_score'].mean(),1),inplace=True)
```

```python
data['character'].fillna('unknown',inplace=True)
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.duplicated().sum()
```

```python
data.dtypes
```

```python
data.describe()
```

as we can see dataset is clean and ready for futher analysis ,the manipulation performed on gviven data are as follows,
1.dataloading and merging:two  data sets,titles.csv and credits.csv,were loaded and merged based on the common column 'id'.
2.handling missing values
.null values in the "description " and "imdb_id" columns were dropped .
.null values in 'age_certification' were replaced with the mode 
.null values in 'seasons','imdb_votes',and 'character' were replaced with  0,0, and ' unknown',respectively.
.null values in 'imdb_score','tmdb_popularity', and 'imdb_score' wre replaced with their respective means 
3.Handling duplicates: duplicate rows identified and removed from the dataset
